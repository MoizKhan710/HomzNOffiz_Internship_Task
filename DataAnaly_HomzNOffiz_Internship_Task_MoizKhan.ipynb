## Handling columns having NAN values

### 1.Price Column

```python
df['Price'] = df.groupby('Rooms')['Price'].apply(lambda x: x.fillna(x.mode()[0])).astype(int)
```

#### Here i am filling the Null values in the Prize column using the Rooms Columns. Basic assumption here is i have grouped the unique values in the rooms column and the prize column then took the mode of each group and filled the prize value accordingly.

### 2.Distance and Postcode

#### Both of these columns have one Null value each. I cannot drop the rows as i would loose a record to instead i will fill these with a '0' value wich would indicate it as unknown information.

```python
df['Distance'] = df['Distance'].fillna(0).astype(int)
df['Postcode'] = df['Postcode'].fillna(0).astype(int)
```

### 3.Bedroom2 Column

```python
df['Bedroom2'] = df.groupby('Rooms')['Bedroom2'].apply(lambda x: x.fillna(x.mode()[0])).astype(int)
```

#### I have grouped the unique values in the rooms column and the Bedroom2 column then took the mode of each group and filled the prize value accordingly.

### 4.Bathroom Column

```python
df['Bathroom'] = df.groupby('Rooms')['Bathroom'].apply(lambda x: x.fillna(x.mode()[0])).astype(int)
```

#### I have grouped the unique values in the rooms column and the Bathroom column then took the mode of each group and filled the prize value accordingly.

### 5.Car Column

```python
df['Car'] = df.groupby('Rooms')['Car'].apply(lambda x: x.fillna(x.mode()[0])).astype(int)
```

#### I have grouped the unique values in the rooms column and the Car column then took the mode of each group and filled the prize value accordingly. Altough car parking space cannot be dertermined with this asumption i have still used it instead of putting zero values.

### 6.Landsize, BuildingArea,YearBuilt  Column

```python
df['Landsize'] = df['Landsize'].fillna(0).astype(int)
df['BuildingArea'] = df['BuildingArea'].fillna(0).astype(int)
df['YearBuilt'] = df['YearBuilt'].fillna(0).astype(int)
```

#### Any values cannot be filled in these places so i will fill '0' value to indicate information is unknown.

### 7.CouncilArea,Regionname,Propertycount Column

```python
df['CouncilArea'].ffill(axis = 0,inplace=True)
df['Regionname'].ffill(axis = 0,inplace=True)
df['Propertycount'] = df['Propertycount'].ffill(axis = 0).astype(int)
```

#### Simply forward filling the values as these columns are sorted

### 8. Lattitude and Longtitude Column

```python
df['Lattitude'].fillna(0,inplace=True)
df['Longtitude'].fillna(0,inplace=True)
```

#### Any values cannot be filled as these are the co-ordinates of a location so filling '0'.

```python
sales_counts = df['Method'].value_counts()
plt.figure(figsize=(10, 6))
sales_counts.plot(kind='bar')
plt.title('Sales Status')
plt.xlabel('Sales Categories')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

#### This is the bar plot which indicates Sales Status of the properties.Here Majority of the properties lie in Sold, Sold Prior and passed in category.


S - property sold;
SP - property sold prior;
PI - property passed in;
PN - sold prior not disclosed;
SN - sold not disclosed;
NB - no bid;
VB - vendor bid;
W - withdrawn prior to auction;
SA - sold after auction;
SS - sold after auction price not disclosed.
N/A - price or highest bid not available.

```python
df['Date'] = pd.to_datetime(df['Date'])
sales_counts = df.groupby(['Date', 'Method']).size().unstack()
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_counts,marker='o', linestyle='-',dashes=False)
plt.title('Sales Status over Time')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

#### This is a lineplot which shows the frequency to different Methods (S - property sold; SP - property sold prior; PI - property passed in,etc) with respect to time. This plot tells us that majority of the property is been sold and least has been withdrawn prior to auction.

```python
df['Month'] = df['Date'].dt.to_period('M')
sales_counts = df.groupby(['Month', 'Method']).size().unstack()
plt.figure(figsize=(12, 6))
sales_counts.plot(marker='o', linestyle='-')
plt.title('Sales Status over Months')
plt.xlabel('Months')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend(title='Method')
plt.show()
```

#### Similarly this line plot tells us the same status in monthly format.

```python
df['Year'] = df['Date'].dt.to_period('Y')
sales_counts = df.groupby(['Year', 'Method']).size().unstack()
plt.figure(figsize=(12, 6))
sales_counts.plot(marker='o', linestyle='-')
plt.title('Sales Status over Years')
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Method')
plt.show()
```

#### And this line plot tells us the same status in yearly format.
